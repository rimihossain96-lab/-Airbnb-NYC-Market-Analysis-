# Data Visualization and Story Telling

## Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables

### Chart - 1

```python
#1. Room Type Distribution
plt.figure(figsize=(10,6))
sns.countplot(x='room_type', data=df)
plt.xlabel("Room Type", fontsize=14, fontstyle="oblique", fontweight="bold", color="olive")
plt.ylabel("Number of Listings", fontsize=14, fontstyle="oblique", fontweight="bold", color="orange")
plt.title("Distribution of Listings by Room Type", fontsize=20, fontstyle="italic", fontweight="bold", color="blue")

plt.xticks(rotation=45)
plt.show()
```

## Why this chart?
Bar chart is best for categorical comparison.

## Insights found?
One room type dominates total listings.

## Business impact?
Helps focus marketing on high-demand room types.

### Chart - 2

```python
#2. Neighbourhood Group Distribution
plt.figure(figsize=(10,6))
sns.countplot(x='neighbourhood_group', data=df)
plt.xlabel("Neighbourhood Group", fontsize=14, fontstyle="oblique", fontweight="bold", color="olive")
plt.ylabel("Number of Listings", fontsize=14, fontstyle="oblique", fontweight="bold", color="orange")
plt.title("Listings Distribution by Neighbourhood Group", fontsize=20, fontstyle="italic", fontweight="bold", color="blue")
plt.xticks(rotation=45)
plt.show()
```

### Chart - 12

```python
import numpy as np

# Select numerical columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Remove infinite values if any
numeric_df = numeric_df.replace([np.inf, -np.inf], np.nan)

# Compute correlation matrix
corr_matrix = numeric_df.corr()

plt.figure(figsize=(12,8))

sns.heatmap(corr_matrix,
            cmap='coolwarm',
            annot=True,
            fmt=".2f",
            linewidths=0.5)

plt.title("Correlation Heatmap: Relationship Between Numerical Variables",
          fontsize=20, fontstyle="italic", fontweight="bold", color="blue")

plt.show()
```

## Why this chart?
A heatmap is used to visualize the strength and direction of relationships between multiple numerical variables at once.

## Insights found?
It shows correlations between price, number of reviews, availability, minimum nights, and other numeric variables.
Strong positive or negative values indicate meaningful relationships.

## Business impact?
Yes, it helps identify key factors affecting pricing, demand, and availability.
Weak or negative correlations may indicate inefficiencies or areas needing strategic improvement.
